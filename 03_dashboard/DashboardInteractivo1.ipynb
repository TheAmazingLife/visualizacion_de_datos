```python
from dash import Dash, dcc, Output, Input  # pip install dash
import dash_bootstrap_components as dbc    # pip install dash-bootstrap-components
```

```python
# Declarar y construir componentes
app = Dash(__name__, external_stylesheets=[dbc.themes.SOLAR])
mytext = dcc.Markdown(children='')
myinput = dbc.Input(value="# Hola. Segundo dashboard en Dash")
```

```python
# Personalizar layout
app.layout = dbc.Container([mytext, myinput])
```

```python
# Callback permite la interacción entre los componentes definidos
@app.callback(
    Output(mytext, component_property='children'),
    Input(myinput, component_property='value')
)

def update_title(user_input):  # los argumentos de la función desde la component_property del Input
    return user_input  # los objetos retornados son asignados a la component_property del Output
```

```python
# Ejecución de la app
if __name__=='__main__':
    app.run_server(port=8052)
```
